```python
import random
import time

class Account: #Q1. Account 클래스

    bank_count = 0 # Q2.클래스 변수
    customer = []

    def __init__(self, name, initial_balance):
        self.bank = None
        self.name = name
        self.number = None
        self.balance = initial_balance
        self.deposit_count = 0
        self.deposit_detail = []
        self.withdraw_detail = []
        #Account.customer.append([self.name,self.balance])
        #self.customer.append(self.bank,self.name,self.number,self.balance,self.deposit_count)

        
    def openAccount(self): #계좌 개설 메서드
        Account.bank_count +=1
        
        self.bank = "SC은행"
        self.number = str(random.randint(100,999)) +"-"+str(random.randint(10,99)) +"-" + str(random.randint(100000,999999))
        print(f"은행이름 : {self.bank}, 계좌번호 : {self.number} 이름 : {self.name}, 잔고 : {self.balance}")
       
    
    def get_account_num(self): # Q3.생성된 계좌의 개수를 출력하는 메서드
        print(f"계좌 갯수는 {Account.bank_count} 입니다")
    
    def deposit(self, balance): # Q4. 입금 메서드 , Q7. 이자 지급하기
        self.deposit_count +=1
        print(f"현재 잔액은 {self.balance} 입니다.")
        self.balance += balance
        print(f"입금 후 잔액은 {self.balance} 입니다.")
        t = time.localtime()
        date = f"{t.tm_year}-{t.tm_mon:02d}-{t.tm_mday:02d} {t.tm_hour:02d}:{t.tm_min:02d}:{t.tm_sec:02d}"
        self.deposit_detail.append({"입금시간":date,"입금액":balance,"잔고":self.balance})

        if self.deposit_count == 5 :
            print(f"이자 지급전 잔고 : {self.balance}")
            self.balance *= 1.01
            print(f"입금 횟수가 5회가 되어 1% 이자가 잔고에 추가되었습니다. 잔고 : {self.balance}")

        

    def withdraw(self, balance): #Q5. 출금 메서드
        print(f"현재 잔액은 {self.balance}입니다")

        if balance <= self.balance:
            self.balance -= balance
        else:
            print("계좌의 잔고 이상으로 출금할 수 없습니다")

        print(f"출금 후 잔액은 {self.balance}입니다")
        
        t = time.localtime()
        date = f"{t.tm_year}-{t.tm_mon:02d}-{t.tm_mday:02d} {t.tm_hour:02d}:{t.tm_min:02d}:{t.tm_sec:02d}"
        self.withdraw_detail.append({"출금시간":date,"출금액":balance,"잔고":self.balance})



    def display_info(self): # Q6. 정보 출력 메서드
        #(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
        print(f"(은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.number}, 잔고: {self.balance:,}원)")
        

    def deposit_history(self): # Q9 입금 내역
        for i in self.deposit_detail:
            print(f"{i}")

    def withdraw_history(self): # Q9 출금 내역
        for i in self.withdraw_detail:
            print(f"{i}")
    

```

```python
#Q8. 여러 객체 생성 (맞는지 모르겠어요...)
a1 = Account("전해진",10000)
b1 = Account("김도하", 1000000)
c1 = Account("파이썬", 20000)

customer = [a1,b1,c1]

a1.deposit(10000)
b1.deposit(2000000)
c1.deposit(20000)
```

```python
# Q9. 특정 고객의 정보 출력 (맞는지 모르겠습니다.....)
for i in customer:
    if i.balance >= 1000000:
        print(f"이름 :{i.name}, 잔고 :{i.balance}")
```
